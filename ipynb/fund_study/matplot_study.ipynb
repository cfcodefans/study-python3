```python
import matplotlib.pyplot as plt
import matplotlib.axes as _ax
import matplotlib.figure as _fig
plt.rcdefaults()    
import numpy as np
fig: _fig.Figure = None
ax: _ax.Axes = None

#nrows=2, ncols=3, it makes 6 charts arranged in 2 rows and 3 columns for each row
fig, ax = plt.subplots()

fig.set_size_inches(18, 4.5)

print(type(fig), fig)
print(type(ax), ax)

xd: np.ndarray = np.arange(0.01, 10.0, 0.01)

yd1: np.ndarray = np.exp(xd)
ax1: _ax.Axes = ax
print(type(ax1.figure), ax1.figure)
ax1.plot(xd, yd1, color='r')
ax1.set_xlabel('time(s)')
ax1.set_ylabel('exp')

yd2: np.ndarray = np.sin(1 * np.pi * xd)
ax2: _ax.Axes = ax1.twinx()
print(ax2.yaxis.get_scale())
print(type(ax2.figure), ax2.figure)
ax2.plot(xd, yd2, color='g')
ax2.set_ylabel('sin')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import pandas as pd
from pandas import DataFrame, Series

_x: np.ndarray = np.arange(0.01, 4.0, 0.01)
_y: np.ndarray = _x * 0.75
df: DataFrame = DataFrame()
df['x'] = _x.tolist()
df['y'] = _y.tolist()
df
fig: Figure = plt.figure(figsize=(8, 6))
ax: Axes = fig.add_axes([0, 0, 4, 3])
ax.plot(df['x'], df['y'])
ax.set_xlabel('x')
ax.set_ylabel('y')
print(ax)
plt.show()
```

```python
from bokeh.io import show, output_file
from bokeh.models import FactorRange
from bokeh.plotting import figure

output_file("bar_mixed.html")

factors = [
    ("Q1", "jan"), ("Q1", "feb"), ("Q1", "mar"),
    ("Q2", "apr"), ("Q2", "may"), ("Q2", "jun"),
    ("Q3", "jul"), ("Q3", "aug"), ("Q3", "sep"),
    ("Q4", "oct"), ("Q4", "nov"), ("Q4", "dec"),

]

p = figure(x_range=FactorRange(*factors), plot_height=350,
           toolbar_location=None, tools="")

x = [ 10, 12, 16, 9, 10, 8, 12, 13, 14, 14, 12, 16 ]
p.vbar(x=factors, top=x, width=0.9, alpha=0.5)

p.line(x=["Q1", "Q2", "Q3", "Q4"], y=[12, 9, 13, 14], color="red", line_width=2)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```
